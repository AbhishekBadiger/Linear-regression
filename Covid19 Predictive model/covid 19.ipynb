```python
#linear regression
from sklearn.model_selection import train_test_split
```

```python
maha
```

```python
#converting date-time to ordinal
maha['date']=maha['date'].map(dt.datetime.toordinal)
maha.head()
```

```python
#getting dependent variable and independent variable
X = maha['date']
y = maha['confirmed']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  #object of the linear resgression
regressor.fit(np.array(X_train).reshape(-1,1),np.array(y_train).reshape(-1,1))    #connects the method which connects the training set
```

```python
maha.tail()
```

```python
y_pred = regressor.predict(np.array([[737930]]))
```

```python
y_pred
```

```python
plt.scatter(np.array(X_train).reshape(-1,1),np.array(y_train).reshape(-1,1), color = 'red')
plt.plot(np.array(X_train).reshape(-1,1), regressor.predict(np.array(X_train).reshape(-1,1)), color = 'blue')  #we dont want the actual value rahter we want the predicted
plt.title('Date vs Confirmed (Training set)')              #hence we have used regressor.predict(X_train)
plt.xlabel('Confirmed')     # here we can see the difference between real salary and predicted salary
plt.ylabel('Date')
plt.show()
```

```python
plt.scatter(np.array(X_test).reshape(-1,1),np.array(y_test).reshape(-1,1), color = 'red')
plt.plot(np.array(X_train).reshape(-1,1), regressor.predict(np.array(X_train).reshape(-1,1)), color = 'blue')  #here we dont have to change because only one regression line 
plt.title('Date vs Confirmed (Test set)')        
plt.xlabel('Confirmed')     
plt.ylabel('Date')
plt.show()
```

```python
from scipy.stats import zscore
maha['Zscore'] = zscore(maha.confirmed)
data = maha[(maha.Zscore > 2.0) | (maha.Zscore < -2.0)]
```

```python
#getting dependent variable and independent variable
X = data['date']
y = data['confirmed']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  #object of the linear resgression
regressor.fit(np.array(X_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
y_pred = regressor.predict(np.array([[737930]]))
```

```python
y_pred
```

```python
plt.scatter(np.array(X_train).reshape(-1,1),np.array(y_train).reshape(-1,1), color = 'red')
plt.plot(np.array(X_train).reshape(-1,1), regressor.predict(np.array(X_train).reshape(-1,1)), color = 'blue')  #we dont want the actual value rahter we want the predicted
plt.title('Date vs Confirmed (Training set)')              #hence we have used regressor.predict(X_train)
plt.xlabel('Confirmed')     # here we can see the difference between real salary and predicted salary
plt.ylabel('Date')
plt.show()
```
